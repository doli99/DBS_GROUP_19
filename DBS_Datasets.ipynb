```python
import pandas as pd

# Load the CNBC dataset
cnbc_data = pd.read_csv('cnbc_headlines.csv')
pd.set_option('display.max_columns', None)
pd.set_option('max_colwidth', None)

# Display the initial structure and information of the dataset to understand what needs cleaning
cnbc_data.info(), cnbc_data.head()
```

```python
# Drop rows with missing values in critical columns
combined_df.dropna(subset=['Headlines', 'Time'], inplace=True)

# Convert 'Time' to datetime format for easier manipulation later on
combined_df['Time'] = pd.to_datetime(combined_df['Time'], errors='coerce')

# Remove any rows where the date could not be parsed
combined_df.dropna(subset=['Time'], inplace=True)

# Optionally, strip any leading or trailing whitespace characters from text columns
combined_df['Headlines'] = combined_df['Headlines'].str.strip()
combined_df['Description'] = combined_df['Description'].str.strip()
```

```python
import re

# Function to clean text data
def clean_text(text):
    # Remove non-alphanumeric characters
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    # Convert to lower case
    text = text.lower()
    return text

# Apply the cleaning function to headlines and descriptions
combined_df['Headlines'] = combined_df['Headlines'].apply(clean_text)
combined_df['Description'] = combined_df['Description'].apply(clean_text)
```

```python
# Mapping of categories to keywords
categories = {
    'General Market': ['market', 'stocks', 'shares', 'investment', 'trading', 'equity', 'securities', 'bonds', 'ETFs', 'bull market', 'bear market', 'market trends', 'volatility', 'index', 'futures'],
    'Financial Instruments': ['derivatives', 'options', 'commodities', 'forex', 'cryptocurrency', 'hedge funds', 'mutual funds', 'pension funds'],
    'Economic Indicators': ['GDP', 'inflation', 'employment', 'interest rates', 'monetary policy', 'fiscal policy', 'economic growth', 'recession', 'consumer confidence'],
    'Investor Behavior': ['investor sentiment', 'retail investors', 'institutional investors', 'market sentiment', 'risk tolerance', 'portfolio management', 'asset allocation'],
    'Specific Events and Policies': ['central bank meetings', 'earnings reports', 'mergers and acquisitions', 'regulatory changes', 'trade agreements', 'economic sanctions', 'geopolitical tensions'],
    'Financial Sectors and Entities': ['banks','bank', 'insurances','real estate', 'FinTech', 'fintech','cryptocurrency','tech stocks', 'energy sector', 'automotive sector', 'healthcare sector','car','plane']
}

# Create a regex pattern for each category by combining keywords
category_patterns = {key: re.compile('|'.join(r'\b{}\b'.format(re.escape(word)) for word in words), re.IGNORECASE) for key, words in categories.items()}
```

```python
# Function to categorize rows based on keywords
def categorize_text(text, patterns):
    matched_categories = []
    for category, pattern in patterns.items():
        if pattern.search(text):
            matched_categories.append(category)
    return ', '.join(matched_categories) if matched_categories else 'Other'

# Apply categorization to the 'Headlines' column
combined_df['Category'] = combined_df['Headlines'].apply(lambda x: categorize_text(x, category_patterns))
```

```python
# Extend the existing categories dictionary with new keywords
categories.update({
    'Legal and Regulatory': ['regulation', 'law', 'court', 'judge', 'legal', 'compliance','regulator','tax','tariff','verdict'],
    'Real Estate': ['housing', 'real estate', 'construction', 'property'],
    'Energy and Commodities': ['oil', 'commodity', 'energy', 'OPEC','gas'],
    'Political Impact': ['politics', 'government', 'presidential', 'election', 'senate', 'parliament','president','vote','economy','foreign'],
    'Corporate News': ['bankruptcy', 'merger', 'acquisition', 'corporate'],
    'Industry Specific': ['fashion', 'aviation', 'retail', 'automotive', 'technology', 'pharmaceutical'],
    'Foreign':['EU','europe','asia','russia','uk','tourism','singapore','china','asain','countries','european'],
    'Covid':['covid','covid-19','corona','coronavirus'],
    'DBS':['dbs','DBS','dbs bank','dbs bank of singapore']
})

# Re-run the categorization with the updated categories
category_patterns = {key: re.compile('|'.join(r'\b{}\b'.format(re.escape(word)) for word in words), re.IGNORECASE) for key, words in categories.items()}
combined_df['Category'] = combined_df['Headlines'].apply(lambda x: categorize_text(x, category_patterns))

# Re-check category distribution
new_category_counts = combined_df['Category'].str.get_dummies(sep=', ').sum().sort_values(ascending=False)
print(new_category_counts)
```

```python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')

# Load your dataset
# Assuming it is saved and reloaded for sentiment analysis
#data = pd.read_csv('path_to_your_categorized_file.csv')

# Initialize VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Function to calculate sentiment score
def sentiment_score(text):
    return sia.polarity_scores(text)['compound']

# Apply sentiment analysis on Headlines
combined_df['Sentiment'] = combined_df['Headlines'].apply(sentiment_score)

# Calculate average sentiment per category
average_sentiment = combined_df.groupby('Category')['Sentiment'].mean()
average_sentiment
```
